```python
#importing pandas library
import pandas as pd
```

```python
#Reading the data from uploaded dataset & storing it inside df
df=pd.read_csv("/content/Weather Data in India from 1901 to 2017.csv")
df
```

```python
#creating a copy of df
df_copy=df.copy()
df_copy
```

```python
#calculating the average temperature of every year & string it inside new column-avgtemp
df_copy=df_copy.assign(avgtemp=(df_copy['JAN']+df_copy['FEB']+df_copy['MAR']+df_copy['APR']+df_copy['MAY']+df_copy['JUN']+df_copy['JUL']+df_copy['AUG']+df_copy['SEP']+df_copy['OCT']+df_copy['NOV']+df_copy['DEC'])/12)
df_copy
```

```python
#droping the null values from the dataframe
df_copy=df_copy.dropna()
```

```python
#creating a new dataframe using the year & avgtemp
df_new= df_copy[['YEAR','avgtemp']]
df_new
```

```python
#renaming Year as ds & avgtemp as y
df_new=df_new.rename(columns = {'YEAR':'ds','avgtemp':'y'})
```

```python
#checking the information of the dataframe
df_new.info()
```

```python
#converting the int64 data to year type
df_new['ds'] = pd.to_datetime(df_new['ds'],format='%Y')
```

```python
#again checking the information of dataframe
df_new.info()
```

```python
#importing matplotlib library & plotting ds respect to y
import matplotlib.pyplot as plt
plt.figure(figsize = (10,5))
plt.scatter(df_new['ds'],df_new['y'])
```

```python
#importing prophet algorithm
import prophet
op = prophet.Prophet()
op.fit(df_new)
```

```python
#continuing the prediction part
forecast = op.make_future_dataframe(periods = 36,freq = 'M')
forecast = op.predict(forecast)
```

```python
#forecasting the final output
op.plot(forecast,xlabel = 'Year',ylabel = 'Temperature(C)')
plt.title("Temperature Forecasting")
```
