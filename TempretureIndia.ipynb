```python
import pandas as pd
```

```python
df=pd.read_csv("/content/Weather Data in India from 1901 to 2017.csv")
df
```

```python
df_copy=df.copy()
df_copy
```

```python
df_copy=df_copy.assign(avgtemp=(df_copy['JAN']+df_copy['FEB']+df_copy['MAR']+df_copy['APR']+df_copy['MAY']+df_copy['JUN']+df_copy['JUL']+df_copy['AUG']+df_copy['SEP']+df_copy['OCT']+df_copy['NOV']+df_copy['DEC'])/12)
df_copy
```

```python
df_copy=df_copy.dropna()
```

```python
df_new = df_copy[['YEAR','avgtemp']]
df_new
```

```python
df_new=df_new.rename(columns = {'YEAR':'ds','avgtemp':'y'})
```

```python
df_new.info()
```

```python
df_new['ds'] = pd.to_datetime(df_new['ds'],format='%Y')
```

```python
df_new.info()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10,5))
plt.scatter(df_new['ds'],df_new['y'])
```

```python
import prophet
op = prophet.Prophet()
op.fit(df_new)
```

```python
forecast = op.make_future_dataframe(periods = 36,freq = 'M')
forecast = op.predict(forecast)
```

```python
op.plot(forecast,xlabel = 'Year',ylabel = 'Tempreture(C)')
plt.title("Tempreture Forecasting")
```
